```python
import re
import os
from pdf2image import convert_from_path
import pytesseract

def extract_invoice_numbers_from_pdf(pdf_path):
    """
    Extract invoice numbers from a given PDF using OCR.

    Args:
        pdf_path (str): The file path to the invoice PDF.

    Returns:
        list of str: A list of extracted invoice numbers (could be empty if none found).
    """
    # Convert PDF to a list of images
    # Each page is converted into one PIL Image object
    images = convert_from_path(pdf_path)

    invoice_numbers = []

    # Simple regex pattern to match something like:
    #   - "Invoice #: 12345"
    #   - "Invoice No: 123-XYZ"
    #   - "Invoice Number: ABC-1234"
    # Adjust this pattern as needed to match your invoice's format.
    invoice_pattern = re.compile(r'(?:invoice\s*no\.?\s*[:#-]?\s*|invoice\s*number\s*[:#-]?\s*)([A-Za-z0-9-]+)', re.IGNORECASE)

    for idx, image in enumerate(images):
        ocr_result = pytesseract.image_to_string(image)
        print(f"--- OCR Result for Page {idx+1} ---")
        print(ocr_result)
        print("-----------------------------\n")

    return invoice_numbers

if __name__ == "__main__":
    # Example usage:
    pdf_file_path = "/Users/cooperfoster/Desktop/output/mayer_multi.pdf"
    
    results = extract_invoice_numbers_from_pdf(pdf_file_path)
    
    # Print out the invoice numbers found
    # This is the only output - as requested
    for invoice_num in results:
        print(invoice_num)
```
